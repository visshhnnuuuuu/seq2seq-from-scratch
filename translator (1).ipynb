```python
special_tokens = [
    "<PAD>",
    "<UNK>"
]
```

```python
MAX_VOCAB = 20000

hindi_words = [
    word
    for word, _ in hindi_counter.most_common(MAX_VOCAB)
]

hindi_vocab = {
    token: idx
    for idx, token in enumerate(
        special_tokens + hindi_words
    )
}
```

```python
def numericalize(sentence, vocab):
    return [
        vocab.get(word, vocab["<UNK>"])
        for word in sentence.split()
    ]
```

```python
from torch.utils.data import Dataset

class TranslationDataset(Dataset):

    def __init__(self, df, hindi_vocab, english_vocab):
        self.df = df
        self.hindi_vocab = hindi_vocab
        self.english_vocab = english_vocab

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):

        hindi_sentence = self.df.iloc[idx]["hindi"]
        english_sentence = self.df.iloc[idx]["english"]

        hindi_ids = numericalize(
            hindi_sentence,
            self.hindi_vocab
        )

        english_ids = numericalize(
            english_sentence,
            self.english_vocab
        )

        return hindi_ids, english_ids
```

```python
dataset = TranslationDataset(
    df,
    hindi_vocab,
    english_vocab
)
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence

PAD_IDX_HINDI = hindi_vocab["<PAD>"]
PAD_IDX_ENGLISH = english_vocab["<PAD>"]

def collate_fn(batch):

    hindi_sentences = []
    english_sentences = []

    for hindi_ids, english_ids in batch:

        hindi_sentences.append(
            torch.tensor(hindi_ids, dtype=torch.long)
        )

        english_sentences.append(
            torch.tensor(english_ids, dtype=torch.long)
        )

    hindi_padded = pad_sequence(
        hindi_sentences,
        batch_first=True,
        padding_value=PAD_IDX_HINDI
    )

    english_padded = pad_sequence(
        english_sentences,
        batch_first=True,
        padding_value=PAD_IDX_ENGLISH
    )

    return hindi_padded, english_padded
```

```python
from torch.utils.data import DataLoader

dataloader = DataLoader(
    dataset,
    batch_size=8,
    shuffle=True,
    collate_fn=collate_fn
)
```

```python
import torch
import torch.nn as nn

class Encoder(nn.Module):

    def __init__(
        self,
        input_dim,
        emb_dim,
        hidden_dim
    ):
        super().__init__()

        self.embedding = nn.Embedding(
            input_dim,
            emb_dim
        )

        self.lstm = nn.LSTM(
            emb_dim,
            hidden_dim,
            batch_first=True
        )

    def forward(self, x):

        embedded = self.embedding(x)

        outputs, (hidden, cell) = self.lstm(
            embedded
        )

        return outputs, hidden, cell
```

```python
class Attention(nn.Module):

    def __init__(self, hidden_dim):
        super().__init__()

        self.attn = nn.Linear(
            hidden_dim * 2,
            hidden_dim
        )

        self.v = nn.Linear(
            hidden_dim,
            1,
            bias=False
        )

    def forward(
        self,
        hidden,
        encoder_outputs
    ):

        src_len = encoder_outputs.shape[1]

        hidden = hidden.squeeze(0)

        hidden = hidden.unsqueeze(1).repeat(
            1,
            src_len,
            1
        )

        energy = torch.tanh(
            self.attn(
                torch.cat(
                    (hidden, encoder_outputs),
                    dim=2
                )
            )
        )

        attention = self.v(
            energy
        ).squeeze(2)

        return torch.softmax(
            attention,
            dim=1
        )
```

```python
class Decoder(nn.Module):

    def __init__(
        self,
        output_dim,
        emb_dim,
        hidden_dim,
        attention
    ):
        super().__init__()

        self.embedding = nn.Embedding(
            output_dim,
            emb_dim
        )

        self.lstm = nn.LSTM(
            emb_dim + hidden_dim,
            hidden_dim,
            batch_first=True
        )

        self.fc = nn.Linear(
            hidden_dim,
            output_dim
        )

        self.attention = attention

    def forward(
        self,
        x,
        hidden,
        cell,
        encoder_outputs
    ):

        # x: [batch_size]
        x = x.unsqueeze(1)

        # embedded: [batch_size, 1, emb_dim]
        embedded = self.embedding(x)

        # attention_weights: [batch_size, src_len]
        attention_weights = self.attention(
            hidden[-1],
            encoder_outputs
        )

        # context: [batch_size, 1, hidden_dim]
        context = torch.bmm(
            attention_weights.unsqueeze(1),
            encoder_outputs
        )

        # [batch_size, 1, emb_dim + hidden_dim]
        rnn_input = torch.cat(
            (embedded, context),
            dim=2
        )

        output, (hidden, cell) = self.lstm(
            rnn_input,
            (hidden, cell)
        )

        prediction = self.fc(
            output.squeeze(1)
        )

        return prediction, hidden, cell
```

```python
class Seq2Seq(nn.Module):

    def __init__(
        self,
        encoder,
        decoder
    ):
        super().__init__()

        self.encoder = encoder
        self.decoder = decoder

    def forward(
        self,
        src,
        trg
    ):

        batch_size = trg.shape[0]
        trg_len = trg.shape[1]
        vocab_size = self.decoder.fc.out_features

        outputs = torch.zeros(
            batch_size,
            trg_len,
            vocab_size
        ).to(src.device)

        encoder_outputs, hidden, cell = self.encoder(src)

        x = trg[:, 0]

        for t in range(1, trg_len):

            prediction, hidden, cell = self.decoder(
                x,
                hidden,
                cell,
                encoder_outputs
            )

            outputs[:, t] = prediction

            x = trg[:, t]

        return outputs
```

```python
encoder = Encoder(
    len(hindi_vocab),
    256,
    512
)
attention = Attention(512)

decoder = Decoder(
    len(english_vocab),
    256,
    512,
    attention=attention
)

model = Seq2Seq(
    encoder,
    decoder
)
```

```python
for src, trg in dataloader:

    outputs = model(
        src,
        trg
    )

    print(outputs.shape)

    break
```

```python
criterion = nn.CrossEntropyLoss(
    ignore_index=english_vocab["<PAD>"]
)
```

```python
import torch.optim as optim

optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "cpu"
)

model = model.to(device)
```

```python
num_epochs = 10

for epoch in range(num_epochs):

    model.train()

    epoch_loss = 0

    for src, trg in dataloader:

        src = src.to(device)
        trg = trg.to(device)

        optimizer.zero_grad()

        output = model(
            src,
            trg
        )

        output_dim = output.shape[-1]

        output = output[:, 1:, :]

        trg = trg[:, 1:]

        output = output.reshape(
            -1,
            output_dim
        )

        trg = trg.reshape(-1)

        loss = criterion(
            output,
            trg
        )

        loss.backward()

        optimizer.step()

        epoch_loss += loss.item()

    print(
        f"Epoch {epoch+1}: "
        f"{epoch_loss/len(dataloader):.4f}"
    )
```
